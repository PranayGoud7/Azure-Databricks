#### header = true
without it          →  col1, col2, col3      (generic names)

with header = True  →  order_id, name, date  (actual column names from csv)

#### inferSchema = True
without it          →  every column is String

with inferSchema    →  spark reads data and guesses types

                       order_id → Integer
                       amount   → Double
                       date     → Date

inferSchema scans the entire file to guess types - slows down on large files. In production, define schema manually

show() is a PySpark built-in that prints a plain text table, works everywhere but truncates long values. 

display() is Databricks only, shows an interactive table with full values and supports charts.

Use display() while exploring in notebooks and show() in production code.

```python
#check first n rows
df_orders.show(10)
```

```python
# basic statistics (min, max, mean, count)
df_orders.describe().show()
```

```python
# select specific columns
df_orders.select("order_id", "product", "price").show()
```

```python
# filter rows
df_orders.filter(df_orders.country == "IN").show()
df_orders.filter(df_orders.price > 100).show()
```

```python
# add new column
df_orders.withColumn("Total",df_orders.qty*df_orders.price).show()
```

```python
#rename cloumn
df_orders.withColumnRenamed("qty","quantity").show()
```
