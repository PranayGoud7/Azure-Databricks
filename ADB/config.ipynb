# Configuration: Azure Authentication
Configure Azure service principal credentials and OAuth 2.0 authentication.

```python
# Fetching Azure service principal credentials from Key Vault backed secret scope
# These values are hidden - printing them shows [REDACTED]
client_id     = dbutils.secrets.get(scope="myADB-scope", key="client-id")      # Azure app client ID
client_secret = dbutils.secrets.get(scope="myADB-scope", key="client-secret")  # Azure app client secret
tenant_id     = dbutils.secrets.get(scope="myADB-scope", key="tenant-AD-id")   # Azure Active Directory tenant ID

# Storage account name where data lake containers (bronze/silver/gold) are created
storage_account = "storage4spotifyproject"

# Telling Spark to use OAuth 2.0 for authentication instead of access keys
# OAuth is more secure - uses token-based auth via Azure Active Directory
spark.conf.set(f"fs.azure.account.auth.type.{storage_account}.dfs.core.windows.net", "OAuth")

# Specifying which OAuth provider to use
# ClientCredsTokenProvider means authenticating using client ID + secret (service principal flow)
spark.conf.set(f"fs.azure.account.oauth.provider.type.{storage_account}.dfs.core.windows.net", "org.apache.hadoop.fs.azurebfs.oauth2.ClientCredsTokenProvider")

# Passing the service principal client ID to Spark
# Spark uses this to identify which app is requesting access
spark.conf.set(f"fs.azure.account.oauth2.client.id.{storage_account}.dfs.core.windows.net", client_id)

# Passing the service principal client secret to Spark
# This is the password for the service principal
spark.conf.set(f"fs.azure.account.oauth2.client.secret.{storage_account}.dfs.core.windows.net", client_secret)

# Azure AD endpoint where Spark will go to get the access token
# Token is then used to access the data lake
spark.conf.set(f"fs.azure.account.oauth2.client.endpoint.{storage_account}.dfs.core.windows.net", f"https://login.microsoftonline.com/{tenant_id}/oauth2/token")
```
