# Magic Commands
## 1. %sql
- Used to run SQL queries in a notebook cell.
## 2. %python
- Used to run Python/PySpark code in a notebook cell.
## 3. %md
- Used to write Markdown text, headings, and documentation in notebooks.
## 4. %r
- Used to run R language code in a notebook cell.
## %fs
- Used to work with DBFS (Databricks File System) commands.
